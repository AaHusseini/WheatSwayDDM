Goal is to complete an example of the analysis process for the videos taken on July 20th 2022, to decide on the best way to extract information from the videos 

## create dataframe

Have extracted information from the drone flight, which contains information such as the drone's position atnd tilt, recorded every 200 ms. 

Set times as index, making sure to convert the index from UTC to greenich time

```python

flight_information = pd.read_csv('/Users/ayahusseini/Documents/Database/original_flight_CSV_files/Jul-20th-2022-05-00PM-Flight-Airdata.csv',
                                 parse_dates=['datetime(utc)'],index_col='datetime(utc)')
flight_information.index = flight_information.index + timedelta(hours=1)
flight_information['time(millisecond)'] = pd.to_timedelta(flight_information['time(millisecond)'],unit='ms')
```

```python
flight_information.head()
```

Convert the time(ms) into a time object

Assume for now that changes from `isVideo=0` to `isVideo=1` represent a video starting and changes from `isVideo =1` to `isVideo = 0` represent a video ending.

```python
# find the times at which a video has started/ended recording 
is_video_series = flight_information['isVideo']
# Create a filter 
video_starts = (is_video_series.shift(1) == 0) & (is_video_series == 1)
video_ends = (is_video_series.shift(-1) == 0) & (is_video_series == 1)
```

for ease, label all of the videos (a total of 22) from 1 to 22.=. Do this in a dataframe `flight_information_video` which contains only the flight times where a video is being filmed. 

```python
flight_information['videoID'] = video_starts.cumsum()
flight_information_video = flight_information[is_video_series == 1]
flight_information_video.head()
```

Load the data containing video names then add a `video_filename` attribute to the flight information d

```python
video_names = pd.read_excel('/Users/ayahusseini/Documents/Database/July-20-2023/Jul_20_2022_video_names.xlsx',header = 1)
video_names.index = range(1,23)
video_names.head()
```

```python
# map the video names data to the flight information df 
flight_information_video['video_filename'] = flight_information_video['videoID'].map(video_names['Video name '].to_dict())
```

We seem to be running into an issue because the weather data was recorded at different flight times than the drone data. To get arund this, we will assume that the wind speed doesn't change much over timescales of $\sim 200 \text{ms}$

```python
import numpy as np
```

```python
np.arange(-200,300,100)
```

```python
# goal is to find the closest matches 
weather_data_approximated = weather_data.copy()
t = np.arange(-300,400,100)
for time_diff in t:
    if time_diff == 0:
        pass
    shifted = weather_data.copy()
    shifted['Flight Time'] = shifted['Flight Time']+timedelta(milliseconds=int(time_diff))
    weather_data_approximated = pd.concat((weather_data_approximated,shifted))
weather_data_approximated = weather_data_approximated.sort_values(by='Flight Time')
weather_data_approximated['Flight Time']
```

```python
merged_df = pd.merge(weather_data_approximated,flight_information_video,on='Flight Time')
merged_df.head()
```
